```python
# To find out the Categorical dataset 

categorical_df = df.select_dtypes(object)
categorical_df
```

```python
# converting all the categorical columns into numerical using Label Encoder

from sklearn.preprocessing import LabelEncoder

lab_enc = LabelEncoder()

for column in categorical_df:
    categorical_df[column] = lab_enc.fit_transform(categorical_df[column])
    
categorical_df
```

```python
# to find out the Numerical dataset

numerical_df = df.select_dtypes(int,float)
numerical_df
```

### Now we are have all the dataset in numerical format so now we will add Categorica_df and numerical_df into one dataframe

# Splitting the data

```python
X = combine_df.iloc[:,:-1]
y = combine_df.iloc[:,-1]
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
print(f"X_train is {X_train.shape} and y_train is {y_train.shape}")
```

```python
print(f"X_test is {X_test.shape} and y_test is {y_test.shape}")
```

# Mode Selection and Training

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 15)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
knn.score(X_test, y_test)*100
```

```python
new_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
new_df.head()
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred) / np.sum(confusion_matrix(y_test, y_pred)), annot = True, cmap = 'twilight', fmt = '.2%' );
plt.title('Confusion Matrix')
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values");
```

# Choosing the K Value

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy_rate = []
error_rate = []

for i in range (1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    score = cross_val_score(knn, X, y, cv = 12)
    accuracy_rate.append(score.mean())
    error_rate.append(1-score.mean())
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)

plt.title('Error rate VS K Value')
plt.xlabel('K')
plt.ylabel('Error Rate');
plt.grid()
```

```python
# First a quick comparison to our original K=10

knn = KNeighborsClassifier(n_neighbors = 24)

knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print('WITH K = 24')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test, pred))
```
